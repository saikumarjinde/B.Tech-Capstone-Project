### 12. Create a list with numbers 1 to 10 and sort it in ascending and descending order.

```python
def ascending_order(li):
  for i in range(len(li)):
    for j in range(i+1,len(li)):
      if li[i] > li[j]:
        li[i],li[j] = li[j],li[i]
  print(li)

def descending_order(li):
  for i in range(len(li)):
    for j in range(i+1,len(li)):
      if li[i] < li[j]:
        li[i],li[j] = li[j],li[i]
  return li

li = [2,3,8,1,5,4,6,7,9]
ascending_order(li)
descending_order(li)
```

### 19.Create a dictionary with 5 key pair values. Key would be a grocery item name and value will be its price (any value in the range of 10 to 1000). (take any grocery items) Write a for loop to iterate over a dictionary and print key-value pairs. While printing if the value is greater Rs. 500 then print a line – “this is a costly item.”

```python
grocery_items = {
    'rice' : 450,
    'wheat' : 600,
    'milk' : 800,
    'BRU' : 300,
    'Boost' : 700
}

grocery_items.items()
grocery_items.keys()
grocery_items.values()

for item,price in grocery_items.items():
  print(f"{item} is Rs. {price}")
  if price > 500:
    print(f"{item} is a costly item")
```

### 20. In the dictionary created in above task delete the key value pairs where price is less than average price of all the items in the dictionary.

```python
grocery_items = {
    'rice' : 950,
    'wheat' : 50,
    'milk' : 40,
    'BRU' : 650,
    'Boost' : 120
}

total_sum = 0
for price in grocery_items.values():
  total_sum += price

avg_price = total_sum/len(grocery_items)

keys_to_delete = [key for key,value in grocery_items.items() if value < avg_price]

for key in keys_to_delete:
  del grocery_items[key]

print(grocery_items)
```
